```python
# I will deal with the average of stay per visit. Therefore, I remove unnecessary columns.
data1 = data.iloc[:,[0,8,9,10,11,12]]
data1
```

```python
# change the columns names 
data1.columns = ['country','2018', '2019', '2020', '2021', '2022']
data1
```

```python
#remove some null vaslus 
data2 = data1.iloc[2:,:]
data2
```

```python
# %load ../src/function.py
import pandas as pd

def remove_region(df:pd.DataFrame) -> pd.DataFrame:
	'''
	This function removes enexpected values(region) of the column: 'country' but only' total world' and 
	something of 'other--- ' are remained.

	Input:pd.DataFrame

	Output:pd.DataFrame  


	''' 
	df2 = df.copy()

	remove_regions = {'North America', 'North America', '- of which EU', '- of which EU15', '- of which EU Oth', 'Europe'}

	#for i in len(df):
	#	if df2['country'] in remove_regions:
	#		df2. 
	df2 = df2[~df2['country'].isin(remove_regions )]





	return df2










```

```python
%run -i ../src/function.py

data3 = data2.copy()
remove_region(data3)
```

```python
#replace the null values in 2020
data4 = data3.copy()
data4.loc[:,'2020'] = np.nan
data4
```

```python
#check the data type
data4.dtypes
```

```python
# change the data type of '2020'
data4['2020'] = data4['2020'].astype('float64')
data4.dtypes
```
