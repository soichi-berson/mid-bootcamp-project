```python
import numpy as np
import pandas as pd
```

```python
#download the files
data = pd.read_csv('../data/semi_cleaned/V_S_L_T_by_country2022.csv')
data.head()
```

#### The column: 'is_EU'(yes/No) is added. 

```python
# import countries' names
eu = pd.read_excel('../data/raw/eu_countries.xlsx', engine="openpyxl")
eu
```

```python
#drop np.nan
eu = eu[ eu['Name'].isna()==False ]
eu.head()
```

```python
data1 = data.copy()
data1['is_EU'] = data1["country"].apply(lambda x: 'Yes' if x in eu.values else "No")
data1
```

#### The column: 'English_spoken'(yes/No) is added. 

```python
# import the data
english = pd.read_excel('../data/raw/english_spoken.xlsx', engine="openpyxl")
english
```

```python
english.iloc[40:,:]
```

```python
#choose only counties whose primary language is English and add the column
english_countries = english.iloc[0:44,0]
english_countries.values
```

```python
data2 = data1.copy()
data2['english_spoken'] = data2["country"].apply(lambda x: 'Yes' if x in english_countries.values else "No")
data2
```

#### The column: 'Visa_requirement'(yes/No) is added.

```python
# import the data
visa = pd.read_excel('../data/raw/visa_requirement.xlsx', engine="openpyxl")
visa.head()
```

```python
data5 = data2.copy()
data5['visa_requirement'] = data5["country"].apply(lambda x: 'Yes' if x in visa.values else "No")
data5
```

### The column: 'commonwealth'(yes/No) is added.

```python
# import the data
commonwealth = pd.read_excel('../data/raw/commonwealth.xlsx', engine="openpyxl")
commonwealth.head()
```

```python
data6 = data5.copy()

# Convert the elements to strings
commonwealth_array = np.array(commonwealth['Country'], dtype=str)

# Remove spaces from the elements of the array
commonwealth_array = np.char.strip(commonwealth_array)


data6['commonwealth'] = data6["country"].apply(lambda x: 'Yes' if x in commonwealth_array else "No")
data6.head()
```

### The column: 'population' is added.

```python
# download the file
population = pd.read_excel('../data/raw/population1.xlsx', engine="openpyxl")
population.head()
```

```python
# choose the data
population1 = population.loc[:,['Country Name', '2022']]
population1 = population1.rename(columns={'Country Name':'country', '2022':'population'})
population1 = population1[ population1['population'].isna()==False ]
population1['population'] = population1['population'].astype('int64')
population1 = population1.reset_index(drop=True)
```

```python
#change the unit of population
population3 = population1.copy()
population3['population'] = round(population3['population']/1000000,2)
population3
```

```python
data8 = data6.copy()

merged_data =  pd.merge(data8,population3 , on='country', how='left')
merged_data.head() 
```

```python
merged_data = merged_data.rename(columns={'population':'population_(m)'})
merged_data.head(1)
```

### The column: 'region' and Income Group are added.

```python
# download the file
region_income = pd.read_excel('../data/raw/region_and_income.xlsx', engine="openpyxl")
region_income.head()
```

```python
merged_data2 = merged_data.copy()

#change the column names 
region_income = region_income.rename(columns={'Country Name':'country', 'Region': 'region', 'IncomeGroup': 'income_group'})
# add the data
merged_data3 =  pd.merge(merged_data2,region_income , on='country', how='left')
merged_data3 
```

###  The column: 'GDP' is added.

```python
# download the file
gdp = pd.read_excel('../data/GDP.xlsx', engine="openpyxl")
gdp.head()
```

```python
# choose the data
gdp1 = gdp.loc[:,['country', '2022']]
gdp1 = gdp1.rename(columns={'2022':'gdp'})
gdp1 = gdp1[ gdp1['gdp'].isna()==False ]
gdp1['gdp'] = round(gdp1['gdp'],2)
gdp1 = gdp1.reset_index(drop=True)
```

```python
merged_data4 = merged_data3.copy()


# add the data
merged_data5 =  pd.merge(merged_data4,gdp1 , on='country', how='left')
merged_data5 
```

```python
#save the data
name = '../data/cleaned/data_' + str(2022) + '.csv' 
merged_data5.to_csv(name, index=False)
```
