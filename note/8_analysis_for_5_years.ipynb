```python
import numpy as np
import pandas as pd
```

#### This note creates the data of for 5 years to use Tableau

```python
a18 = pd.read_csv('../data/semi_cleaned/visit_spending_length_total2018.csv')
a19 = pd.read_csv('../data/semi_cleaned/visit_spending_length_total2019.csv')
a21 = pd.read_csv('../data/semi_cleaned/visit_spending_length_total2021.csv')
a22 = pd.read_csv('../data/semi_cleaned/visit_spending_length_total2022.csv')
```

```python
a18.columns = ['country','number_of_visit', 'spending_per_day', 'length_of_stay', 'total_(m)']
a19.columns = ['country','number_of_visit', 'spending_per_day', 'length_of_stay', 'total_(m)']
a21.columns = ['country','number_of_visit', 'spending_per_day', 'length_of_stay', 'total_(m)']
a22.columns = ['country','number_of_visit', 'spending_per_day', 'length_of_stay', 'total_(m)']
```

```python
data_merged = pd.concat([a18, a19], ignore_index=True)
data_merged = pd.DataFrame(data_merged[data_merged['country'] == 'Total World']).reset_index(drop=True)
data_merged.iloc[0,0] = '2018'
data_merged.iloc[1,0] = '2019'
data_merged
```

```python
#create a DataFrame for 2020
# the number of visit is 11101 from the raw data
data_20 = [['2020', 11101, np.nan, np.nan, np.nan]]
df_20 = pd.DataFrame(data_20, columns=['country','number_of_visit', 'spending_per_day', 'length_of_stay', 'total_(m)'])

data_merged1 = pd.concat([data_merged, df_20], ignore_index=True)
data_merged1
```

```python
a21.tail()
```

```python
data_merged3 = pd.concat([a21, a22], ignore_index=True)
#data_merged3 = pd.DataFrame(data_merged[data_merged['country'] == 'Total World']).reset_index(drop=True)
data_merged3 = data_merged3.iloc[[58,117],:]
data_merged3.iloc[0,0] = '2021'
data_merged3.iloc[1,0] = '2022'
data_merged3 = data_merged3.reset_index(drop=True)
data_merged3
```

```python
data_merged4 = pd.concat([data_merged1, data_merged3], ignore_index=True).reset_index(drop=True)
data_merged4
```

```python
# change the column names 
data_merged4.columns = ['year','number_of_visit', 'spending_per_day', 'length_of_stay', 'total_(m)']
data_merged4["year"] = pd.array(data_merged4["year"], dtype=pd.Int64Dtype())

data_merged4.dtypes
```

```python
#save the data
name = '../data/cleaned/5_years_total2.csv' 
data_merged4.to_csv(name, index=False)
```

```python
b18 = pd.read_csv('../data/cleaned/data_2018.csv')
b19 = pd.read_csv('../data/cleaned/data_2019.csv')
b21 = pd.read_csv('../data/cleaned/data_2021.csv')
b22 = pd.read_csv('../data/cleaned/data_2022.csv')
```

```python
b18
```

```python
groupby18 = b18.groupby(['region']).agg({'total_2018_(m)': ['sum'] })
groupby19 = b19.groupby(['region']).agg({'total_2019_(m)': ['sum'] })
groupby21 = b21.groupby(['region']).agg({'total_2021_(m)': ['sum'] })
groupby22 = b22.groupby(['region']).agg({'total_2022_(m)': ['sum'] })
display(groupby18.T)
```

```python
data20 = [[ np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]]
df_20 = pd.DataFrame(data20, columns=['East Asia & Pacific', 'Europe & Central Asia', 'Latin America & Caribbean'
                        , 'Middle East & North Africa', 'North America', 'South Asia', 'Sub-Saharan Africa'])
df_20
```

```python
data_merged5 = pd.concat([groupby18.T, groupby19.T,df_20, groupby21.T,groupby22.T], ignore_index=True)
data_merged5
```

```python
# add year
df = pd.DataFrame({'year': [2018, 2019,2020, 2021, 2022]})
data_merged6 = pd.concat([df, data_merged5], axis = 1, ignore_index=True)
data_merged6
```

```python
data_merged6.columns = ['year','East Asia & Pacific	', 'Europe & Central Asia', 'Latin America & Caribbean'
                        , 'Middle East & North Africa', 'North America', 'South Asia', 'Sub-Saharan Africa']
data_merged6
```
